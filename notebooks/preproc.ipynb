```python
import joblib
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
x_train = joblib.load('../data/processed/x_train.pkl')
y_train = joblib.load('../data/processed/y_train.pkl')
x_test = joblib.load('../data/processed/x_test.pkl')
y_test = joblib.load('../data/processed/y_test.pkl')
x_valid = joblib.load('../data/processed/x_valid.pkl')
y_valid = joblib.load('../data/processed/y_valid.pkl')
```

```python
print(x_train.shape , y_train.shape)
print(x_test.shape , y_test.shape)
print(x_valid.shape , y_valid.shape)
```

```python
x_train.describe()
```

```python
sc = StandardScaler()
x_train_nr = sc.fit_transform(x_train)
```

```python
x_train_norm = pd.DataFrame(x_train_nr, columns=x_train.columns)
x_train_norm
```

```python
new_cols = ['Temperature', 'Humidity', 'TVOC', 'eCO2', 'Raw_H2', 'Raw_Ethanol', 'Pressure', 'PM1.0']

x_test = x_test[new_cols].copy()
x_valid = x_valid[new_cols].copy()
```

```python
joblib.dump(x_train, '../data/processed/x_train.pkl')
joblib.dump(x_train_norm, '../data/processed/x_train_norm.pkl')
joblib.dump(y_train, '../data/processed/y_train.pkl')
joblib.dump(x_valid, '../data/processed/x_valid.pkl')
joblib.dump(y_valid, '../data/processed/y_valid.pkl')
joblib.dump(x_test, '../data/processed/x_test.pkl')
joblib.dump(y_test, '../data/processed/y_test.pkl')
```
